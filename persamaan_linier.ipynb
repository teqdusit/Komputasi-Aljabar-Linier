#PERSAMAAN LINIER

Persamaan linear adalah persamaan aljabar yang variabelnya berpangkat satu dan membentuk garis lurus saat digambarkan dalam grafik. Persamaan ini terdiri dari konstanta dan variabel, umumnya berbentuk
 ax + b =0 (satu variabel) atau y = mx + c
 (dua variabel, di mana m
 adalah kemiringan/gradien)

 Komponen utama:

 - Variabel: Huruf pengganti nilai yang dicari (misal: x, y
).

 - Koefisien: Angka yang berada di depan variabel.

 - Konstanta: Angka tetap yang tidak memiliki variabel

 Jenis jenis persamaan linier:

 1. Persamaan Linear Satu Variabel (PLSV): Memiliki satu variabel, contoh:2x + 4 = 10


 2. Persamaan Linear Satu Variabel (PLSV): Memiliki satu variabel, contoh:
3x + y = 6

 3. Sistem Persamaan Linear (SPL): Sekumpulan persamaan linear dengan variabel sama, contoh:
  - 2x + 3y = 6
 - x - y = 2

 Dua persamaan linier dua variabel:

contoh 1

 - 2x +3y = 12

 - x + y = 11

 contoh 2

 - 4x - 2y = 6

 - 3x + y = 11
